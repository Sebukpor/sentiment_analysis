```python
# ==============================
# IMPORTS
# ==============================
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    classification_report,
    confusion_matrix,
)
from transformers import (
    BertTokenizerFast,
    BertForSequenceClassification,
    Trainer,
    TrainingArguments,
    EarlyStoppingCallback,
    DataCollatorWithPadding,
    pipeline,
)
from datasets import Dataset
import logging
import os
import json
import time
from datetime import datetime
import warnings

warnings.filterwarnings("ignore")

# ==============================
# CONFIGURATION
# ==============================
class Config:
    # Data
    DATA_PATH = "/content/customer_reviews.csv"
    TEXT_COLUMN = "Cleaned_Comments"
    LABEL_COLUMN = "Sentiment"

    # Labels (FIXED – as you requested)
    LABEL_MAPPING = {
        0: "negative",
        1: "neutral",
        2: "positive",
    }

    # Model
    MODEL_NAME = "bert-base-uncased"
    NUM_LABELS = 3
    MAX_LENGTH = 128
    BATCH_SIZE = 16

    # Training
    LEARNING_RATE = 2e-5
    WEIGHT_DECAY = 0.01
    NUM_EPOCHS = 4
    EVAL_STEPS = 500
    SAVE_STEPS = 1000
    PATIENCE = 2

    # Output
    OUTPUT_DIR = "./bert_sentiment_production"
    LOG_DIR = "./logs"
    VERSION = f"v{datetime.now().strftime('%Y%m%d_%H%M%S')}"

    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
    FP16 = torch.cuda.is_available()


# ==============================
# LOGGING
# ==============================
def setup_logging():
    os.makedirs(Config.LOG_DIR, exist_ok=True)
    logging.basicConfig(
        level=logging.INFO,
        format="%(asctime)s | %(levelname)s | %(message)s",
        handlers=[
            logging.StreamHandler(),
            logging.FileHandler(
                os.path.join(Config.LOG_DIR, f"train_{Config.VERSION}.log")
            ),
        ],
    )
    return logging.getLogger("Sentiment")

logger = setup_logging()

# ==============================
# DATA LOADING & VALIDATION
# ==============================
def load_data() -> pd.DataFrame:
    df = pd.read_csv(Config.DATA_PATH)

    df = df.dropna(subset=[Config.TEXT_COLUMN, Config.LABEL_COLUMN])
    df[Config.LABEL_COLUMN] = df[Config.LABEL_COLUMN].astype(int)
    df = df[df[Config.LABEL_COLUMN].isin([0, 1, 2])]

    logger.info("Label distribution:")
    logger.info(df[Config.LABEL_COLUMN].value_counts(normalize=True))

    return df


def split_data(df):
    train_df, eval_df = train_test_split(
        df,
        test_size=0.1,
        stratify=df[Config.LABEL_COLUMN],
        random_state=42,
    )

    train_ds = Dataset.from_pandas(train_df[[Config.TEXT_COLUMN, Config.LABEL_COLUMN]])
    eval_ds = Dataset.from_pandas(eval_df[[Config.TEXT_COLUMN, Config.LABEL_COLUMN]])

    return train_ds, eval_ds


# ==============================
# TOKENIZATION
# ==============================
def tokenize(ds, tokenizer):
    def fn(batch):
        return tokenizer(
            batch[Config.TEXT_COLUMN],
            truncation=True,
            max_length=Config.MAX_LENGTH,
        )

    ds = ds.map(fn, batched=True, remove_columns=[Config.TEXT_COLUMN])
    ds = ds.rename_column(Config.LABEL_COLUMN, "labels")
    ds.set_format("torch")

    return ds


# ==============================
# MODEL
# ==============================
def build_model():
    model = BertForSequenceClassification.from_pretrained(
        Config.MODEL_NAME,
        num_labels=Config.NUM_LABELS,
        id2label=Config.LABEL_MAPPING,
        label2id={v: k for k, v in Config.LABEL_MAPPING.items()},
    )
    model.to(Config.DEVICE)
    return model


# ==============================
# METRICS
# ==============================
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)

    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_weighted": f1_score(labels, preds, average="weighted"),
        "f1_macro": f1_score(labels, preds, average="macro"),
        "precision": precision_score(labels, preds, average="weighted", zero_division=0),
        "recall": recall_score(labels, preds, average="weighted", zero_division=0),
    }


# ==============================
# TRAINING
# ==============================
def train(model, train_ds, eval_ds, tokenizer):
    args = TrainingArguments(
        output_dir=os.path.join(Config.OUTPUT_DIR, "checkpoints"),
        eval_strategy="steps",
        eval_steps=Config.EVAL_STEPS,
        save_steps=Config.SAVE_STEPS,
        save_total_limit=2,
        per_device_train_batch_size=Config.BATCH_SIZE,
        per_device_eval_batch_size=Config.BATCH_SIZE * 2,
        learning_rate=Config.LEARNING_RATE,
        weight_decay=Config.WEIGHT_DECAY,
        num_train_epochs=Config.NUM_EPOCHS,
        fp16=Config.FP16,
        load_best_model_at_end=True,
        metric_for_best_model="eval_f1_macro",
        greater_is_better=True,
        logging_steps=100,
        seed=42,
        report_to="none",
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_ds,
        eval_dataset=eval_ds,
        tokenizer=tokenizer,
        data_collator=DataCollatorWithPadding(tokenizer),
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(Config.PATIENCE)],
    )

    trainer.train()
    metrics = trainer.evaluate()

    logger.info("Final metrics:")
    for k, v in metrics.items():
        logger.info(f"{k:20s} = {v:.4f}")

    return trainer


# ==============================
# PRODUCTION INFERENCE (CALIBRATED)
# ==============================
class SentimentClassifier:
    def __init__(self, model_path):
        self.tokenizer = BertTokenizerFast.from_pretrained(model_path)
        self.model = BertForSequenceClassification.from_pretrained(model_path)

        self.model.config.id2label = Config.LABEL_MAPPING
        self.model.config.label2id = {v: k for k, v in Config.LABEL_MAPPING.items()}

        self.model.to(Config.DEVICE)
        self.model.eval()

        self.pipe = pipeline(
            "text-classification",
            model=self.model,
            tokenizer=self.tokenizer,
            device=0 if Config.DEVICE == "cuda" else -1,
            return_all_scores=True,
        )

    def predict(self, texts, neutral_margin=0.15):
        results = []

        for out in self.pipe(texts):
            scores = {o["label"]: o["score"] for o in out}

            best_label = max(scores, key=scores.get)
            best_score = scores[best_label]

            # 🔧 Neutral calibration (critical)
            if (
                best_label == "positive"
                and abs(scores["positive"] - scores["neutral"]) < neutral_margin
            ):
                best_label = "neutral"

            results.append(
                {
                    "label": best_label,
                    "confidence": best_score,
                    "probabilities": scores,
                }
            )

        return results


# ==============================
# MAIN
# ==============================
def main():
    df = load_data()
    train_ds, eval_ds = split_data(df)

    tokenizer = BertTokenizerFast.from_pretrained(Config.MODEL_NAME)
    train_tok = tokenize(train_ds, tokenizer)
    eval_tok = tokenize(eval_ds, tokenizer)

    model = build_model()
    trainer = train(model, train_tok, eval_tok, tokenizer)

    # ---- Diagnostics (IMPORTANT) ----
    preds = trainer.predict(eval_tok)
    y_true = preds.label_ids
    y_pred = np.argmax(preds.predictions, axis=1)

    logger.info("\nClassification report:")
    logger.info(
        classification_report(
            y_true, y_pred, target_names=["negative", "neutral", "positive"]
        )
    )

    logger.info("\nConfusion matrix:")
    logger.info(confusion_matrix(y_true, y_pred))

    # ---- Save final model ----
    final_path = os.path.join(Config.OUTPUT_DIR, Config.VERSION)
    trainer.save_model(final_path)
    tokenizer.save_pretrained(final_path)

    # ---- Production test ----
    clf = SentimentClassifier(final_path)
    examples = [
        "Great service and clean rooms!",
        "The wifi was slow and breakfast was cold.",
        "Average hotel, nothing special but nothing wrong either.",
    ]

    for t, r in zip(examples, clf.predict(examples)):
        print(t)
        print(r)
        print("-" * 50)


if __name__ == "__main__":
    main()
```
